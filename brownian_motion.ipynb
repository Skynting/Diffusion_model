# Simulating Brownian motion

Euler simulation of SDE of the type $dX = \mu(X) dt + \sigma(X) dW, X_0 = x.$  

In the code below $\mu(x) = \sin(x), \sigma(x)=1$ 

```python
import numpy as np
import matplotlib.pyplot as plt

x0 = 1  # initial value
n = 1000  # number of partition points
t = np.linspace(0, 1, n)  # gridpoints
dt = 1 / (n - 1)  # partition step
N = 10  # number of simulations
x = x0 * np.ones((n, N))
mu = lambda x: np.sin(x)
sigma = lambda x: 1
dW = np.sqrt(dt) * np.random.randn(n, N)

for i in range(n - 1):
    x[i + 1, :] = x[i, :] + mu(x[i, :]) * dt + sigma(x[i, :]) * dW[i, :]

# Plot the simulations
plt.figure(figsize=(10, 6))
for i in range(N):
    plt.plot(t, x[:, i], label=f'Simulation {i + 1}')

plt.xlabel('Time')
plt.ylabel('Value of x')
plt.title('Simulations of Brownian Motion')
plt.legend()
plt.show()
```

## Simulating Brownian bridge

The Brownian bridge is specified by the SDE:
$$
\left\{\begin{array}{l}
d x_t=\frac{X_T-x_t}{1-t} \cdot d t+d W_t \\
X_0=0
\end{array}\right.
$$
where $\left\{W_t\right\}_{t \in [0,1]}$ denotes the wiener process.

In the code below $X_T=10$

```python
import numpy as np
import matplotlib.pyplot as plt

x0 = 0  # initial value
n = 1000  # number of partition points
t = np.linspace(0, 1, n)  # gridpoints
dt = 1 / (n - 1)  # partition step
N = 10  # number of simulations
x = x0 * np.ones((n, N))
mu = lambda x,t: (10-x)/(1-t)
sigma = lambda x: 1
dW = np.sqrt(dt) * np.random.randn(n, N)

for i in range(n - 1):
    x[i + 1, :] = x[i, :] + mu(x[i, :], t[i]) * dt + sigma(x[i, :]) * dW[i, :]

# Plot the simulations
plt.figure(figsize=(10, 6))
for i in range(N):
    plt.plot(t, x[:, i], label=f'Simulation {i + 1}')

plt.xlabel('Time')
plt.ylabel('Value of x')
plt.title('Simulations of Brownian Bridge')
plt.legend()
plt.show()
```
